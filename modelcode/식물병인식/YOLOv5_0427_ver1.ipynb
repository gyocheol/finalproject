```python
## 잎 부위 촬영이 아니면 제외하는 함수 (잎 코드는 03)
def leaf_image(datas):
    remove_list = []
    for data in datas:

        # 잎이 아닌 데이터면 remove_list에 추가
        if data.split('_')[5] != '03':
            remove_list.append(data)

        # 잎 데이터만 new_datas에 추가
        new_datas = [x for x in datas if x not in remove_list]
    
    return new_datas
```

```python
## 흰가루병에 해당하는 사진이 아니면 제외
# 질병코드(index 3): 고추 02, 애호박 08, 오이 12, 호박 20
# 작물코드(index 4): 고추 01, 애호박 04, 오이 06, 호박 10

def powdery_files(files):
    
    # 모든 이미지 불러오기
    for file in files:
        disease = file.split('_')[3]
        crops = file.split('_')[4]

        # 해당 작물 + 흰가루병인 경우만 리스트에 추가
        powdery_disease = []

        if disease == '02' and crops == '01': powdery_disease.append(file)     # 고추
        elif disease == '08' and crops == '04' : powdery_disease.append(file)  # 애호박
        elif disease == '12' and crops == '06' : powdery_disease.append(file)  # 오이
        elif disease == '20' and crops == '10' : powdery_disease.append(file)  # 호박

    return powdery_disease
```

#### 이미지 데이터 Train, Validation, Test 분리

```python
%cd /


# train, validation 데이터는 나뉘어져 있으므로, train 데이터에서 train/test 분리
# 분리 전, train, validation 데이터 크기 확인
img_list = glob('/dataset/Aihub/images/valid/*.jpg')
print(f'train/test 데이터 크기: {len(img_list)}')

val_img_list = glob('/dataset/Aihub/images/valid/*.jpg')
print(f'validation 데이터 크기: {len(val_img_list)}')
```

```python
# Train, Test 데이터 분리

from sklearn.model_selection import train_test_split

train_img_list, test_img_list = train_test_split(img_list, test_size = 0.2, random_state=2000)
print(f'train 데이터 크기: {len(train_img_list)}')
print(f'test 데이터 크기: {len(test_img_list)}')
```

```python
# Train, Validaion 이미지 경로를 txt 파일로 저장

with open('./dataset/train.txt', 'w') as f:
    f.write('\n'.join(train_img_list) + '\n')

with open('./dataset/val.txt', 'w') as f:
    f.write('\n'.join(val_img_list) + '\n')
```

```python
import yaml

with open('/dataset/data.yaml', 'w') as f:
    data = yaml.load(f)

print(data)

data['train'] = '/dataset/train.txt'
data['val'] = '/dataset/val.txt'

with open('/dataset/data.yaml', 'w') as f:
    yaml.dump(data, f)

print(data)
```
